**SIMPLE EXAMPLES**

```python
A = pd.DataFrame({'Transport':['poyezd','avtomobil','kema'],
                  'Tezligi':['70','80','100']})
A
```

```python
B = pd.DataFrame({'Transport':['poyezd','avtomobil','tramvay'],
                  'Narxi':['70000','80900','100900']})
B
```

**MERGE**

```python
merged = pd.merge(A,B,how = "inner") #inner join - ichki
merged
```

```python
merged_outer = pd.merge(A,B,how = "outer") #outer join -tashqi
merged_outer
```

```python
merged_left = pd.merge(A,B,how = "left") #left join - chapgi birlashtirish
merged_left
```

```python
merged_right  = pd.merge(A,B, how = "right") #right join - ongga birlashtirish
merged_right
```

```python
A = pd.DataFrame({'Transport-turi':['poyezd','avtomobil','kema'],
                  'Tezligi':['70','80','100']})
A
```

```python
B
```

```python
merg = pd.merge(A,B,left_on = "Transport-turi",right_on = "Transport")
merg
```

**Indesklar ustida birlashtirish**

```python
A =   A.set_index("Transport-turi")
A
```

```python
B
```

```python
merged1 = pd.merge(A,B,left_index = True, right_on = 'Transport')
merged1
```

```python
B = B.set_index('Transport')
B
```

```python
merged2 = pd.merge(A,B,left_index = True, right_index = True)
merged2
```
